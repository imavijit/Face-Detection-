```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#read the image
face_image=cv2.imread("barcelona.jpg")
plt.figure(figsize=(16,8))
plt.imshow(face_image)
```

```
#convert image to grey image
face_grey=cv2.cvtColor(face_image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(16,8))
plt.imshow(face_grey,cmap='gray')
```

```
#loading the training file
face_clf = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```
face_detect=face_clf.detectMultiScale(face_grey,scaleFactor=1.1,minNeighbors=5)
#number of face found
print("faces found:",len(face_detect))
face_detect
```

```
for (x,y,w,h) in face_detect:
  cv2.rectangle(face_image,(x,y),(x+w,y+h),(0,255,0),2)
```

```
plt.figure(figsize=(16,8))
plt.imshow(cv2.cvtColor(face_image,cv2.COLOR_BGR2RGB))
```
